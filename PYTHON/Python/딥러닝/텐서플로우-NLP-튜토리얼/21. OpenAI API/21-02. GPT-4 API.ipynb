```python
import openai
client = openai.OpenAI(api_key="sk-xxxxxx...")
```

```python
response = client.chat.completions.create(
    model="gpt-5.5",
    messages=[
        {"role": "user", "content": "2020년 월드시리즈에서는 누가 우승했어?"}
    ]
)

# 답변을 받는 방법
print(response.choices[0].message.content)
```
