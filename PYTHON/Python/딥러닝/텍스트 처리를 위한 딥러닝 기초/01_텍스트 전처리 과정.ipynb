# **환경 설정 및 라이브러리 설치**

```python
# 1. 필수 라이브러리 설치
!pip install konlpy kss

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from konlpy.tag import Okt
import kss
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

nltk.download('punkt')
nltk.download('stopwords')
```

# **단계별 실습 코드**

```python
### ## 1단계: Tokenization (Word & Sentence) 기법 수행하기
# 한글, 영어 토크나이저를 사용하고 차이점 연구

nltk.download('punkt_tab') # punkt 데이터 다운로드

# 영어 토큰화 (단어 & 문장)
en_text = "I am actively looking for Ph.D. students. and you are a Ph.D. student."
print("영어 문장 토큰화:", sent_tokenize(en_text))
print("영어 단어 토큰화:", word_tokenize(en_text))

# 한국어 토큰화 (형태소 분석 & 문장)
okt = Okt()
ko_text = "딥 러닝 자연어 처리가 재미있기는 합니다. 그런데 문제는 영어보다 한국어로 할 때 너무 어렵습니다."
print("한국어 형태소 분석(Okt):", okt.morphs(ko_text))
print("한국어 문장 토큰화(KSS):", kss.split_sentences(ko_text))
```

```python
### ## 2단계: Cleaning & Normalization 수행하기
# 전처리 과정을 통해 텍스트 정제 (대소문자 통합 등)

# 영어의 경우 대소문자 통합(Normalizing)이 일반적임
en_tokens_norm = [w.lower() for w in word_tokenize(en_text)]
print("영어 정규화(소문자화):", en_tokens_norm)
```

```python
### ## 3단계: Stopwords 구분방법 이해하기
# 한글/영어 불용어 처리방식 비교분석

# 영어: NLTK에서 제공하는 리스트 사용
stop_words_en = set(stopwords.words('english'))
en_filtered = [w for w in en_tokens_norm if w not in stop_words_en]
print("영어 불용어 제거 결과:", en_filtered)

# 한글: 직접 정의한 불용어 사전 사용 (한글은 조사/어미가 다양하여 직접 지정이 효과적)
stop_words_ko = "가 는 를 에 의 와 과"
stop_words_ko = set(stop_words_ko.split(' '))
ko_filtered = [w for w in okt.morphs(ko_text) if w not in stop_words_ko]
print("한국어 불용어 제거 결과:", ko_filtered)
```

```python
### ## 4단계: Integer Encoding (원핫인코딩)
# 정수인코딩 및 원-핫 인코딩의 컨셉 실습

text_data = ["나는 자연어 처리를 배운다"]
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_data)

# 정수 인코딩
encoded = tokenizer.texts_to_sequences(text_data)[0]
print("단어 정수 인코딩:", encoded)

# 원-핫 인코딩
one_hot = to_categorical(encoded)
print("원-핫 벡터:\n", one_hot)
```

```python
### ## 5단계: Vectorization (DTM, TF-IDF)
# 벡터화 이론 실습 및 결과 확인

corpus = ["먹고 싶은 햄버거", "햄버거가 맛있는 집", "햄버거 최고야"]

# DTM (문서 단어 행렬)
dtm_vector = CountVectorizer()
print("DTM 결과:\n", dtm_vector.fit_transform(corpus).toarray())

# TF-IDF (단어 빈도-역문서 빈도)
tfidf_vector = TfidfVectorizer()
print("TF-IDF 결과:\n", tfidf_vector.fit_transform(corpus).toarray())
```
